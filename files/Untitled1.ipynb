```python
f = open("Missions.csv","r")
data = f.readlines()
f.close()

years = [int(x.split(",")[2].split()[2]) for x in data[1:]]

outcomes = [x.split(",")[7] for x in data[1:]]
success = outcomes.count("Successful")/len(outcomes)
country = [x.split(",")[4] for x in data[1:]]
countryCount = [country.count(x) for x in set(country)]
print(set(country))
print(countryCount)
print(country)
print(success)
counts = [years.count(x) for x in set(years)]
print(counts)

from matplotlib import pyplot as plt

#plt.plot(list(set(years)),counts, 'r.')
plt.figure(figsize = (40,40))

plt.bar(range(len(set(country))),countryCount,tick_label = list(set(country)))
plt.xticks(fontsize = 32)
plt.yticks(fontsize = 32)
```

```python
from numpy import linspace,meshgrid,sqrt
from matplotlib import pyplot as plt
from matplotlib import cm

x = linspace(-5,10,100)
y = linspace(-3,8,100)

X,Y = meshgrid(x,y)

k = 8.99e9
e = 1.602e-19

V = k * e * (1/sqrt((X-5)**2 + Y**2) + 1/sqrt(X**2 + (Y-5)**2) + 1/sqrt(X**2 + Y**2) )

plt.figure(figsize = (30,30))
ax = plt.axes(projection = "3d")
ax.plot_surface(X,Y,V)
ax.set_zlim(0,7.5e-9)
```
